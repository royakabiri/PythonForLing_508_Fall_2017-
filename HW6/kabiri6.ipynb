<li>Read in the <code>alice.txt</code> file and:
<ol>
<li>Strip header, convert punctuation to spaces, convert to lowercase, and split into words.
<li>Count the words and print that value.
<li>Do lexical statistics on the odd-numbered words and report back the 20 most frequent along with their counts.
<li>Do lexical statistics on the <em>even</em>-numbered words and report back the 20 most frequent along with their counts.
</ol>

```python
import re
#rk:read the text file
alice = open('alice.txt','r')
aliceText = alice.read()
alice.close()
aliceText = aliceText[10900:]

#rk:convert to lower case
lowerAliceText = aliceText.lower()
#rk:replace all punctuation with spaces
newAliceText= re.sub('[\.\?\-!\?\*,"\'(\):\`\[\];_/~]',' ',lowerAliceText)
#print(newAliceText)

#rk:split into words
aliceWords = newAliceText.split()
#print(aliceWords)
print ('Number of words in Alice text: ', len(aliceWords),'\n')

#rk:creating two empty list to store odd and even words
oddList=[]
evenList=[]
count = 1
for word in aliceWords:
    if count % 2 == 0:
        evenList.append(word)
    else:
        oddList.append(word)
    count += 1
    
#print(oddList)
#print(evenList)
#print(len(oddList))
#print(len(evenList))

#rk:count frequency of odd words
oddWordDict=  {}
for w in oddList:
    if w in oddWordDict:
         oddWordDict[w]+=1
    else:
         oddWordDict[w]= 1
#print (oddWordDict)

#rk:count frequency of even words
evenWordDict=  {}
for w in evenList:
    if w in evenWordDict:
         evenWordDict[w]+=1
    else:
         evenWordDict[w]= 1
#print (evenWordDict)  

#rk:sort oddWordDict which makes it as a list of tuples  
sortedOddList = sorted(oddWordDict.items(), key=lambda x: x[1], reverse=True)
#print (sortedOddList)

#rk:sort evenWordDict which makes it as a list of tuples  
sortedEvenList = sorted(evenWordDict.items(), key=lambda x: x[1], reverse=True)
#print (sortedEvenList)

#rk:convert sorted odd list back to a dictionary
finalOddDict = dict(sortedOddList)
#print (finalOddDict) 

#rk:convert sorted even list back to a dictionary
finalEvenDict = dict(sortedEvenList)
#print (finalEvenDict) 

#rk:Print out the 10 most frequent odd words and their counts.
print ("Twenty most frequent odd words are as follows: ")  
count = 1
for key in finalOddDict:
    if count < 21:
           print (count, key, ":", finalOddDict[key],)
    count += 1
print('\n')

#rk:Print out the 10 most frequent even words and their counts.
print ("Twenty most frequent even words are as follows: ")  
count = 1
for key in finalEvenDict:
    if count < 21:
           print (count, key, ":", finalEvenDict[key])
    count += 1
```
