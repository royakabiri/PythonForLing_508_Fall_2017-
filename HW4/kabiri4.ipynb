```python
import re

lines = []
f = open('alice.txt','r')
t = f.read()
f.close()

words = t.lower().split()
#print (words)
pats = []
#matches any word that begins and ends with the same letter    
pats.append('^(\w)\w*\\1$') #or r'^([a-z])\[a-z]*\1$'

#matches a word with an odd number of letters          
pats.append ('^[a-z]([a-z]{2})*$') #or:^[a-z]([a-z][a-z])*$    ,for even numbers: '^([a-z][a-z])+$' or '^([a-z]{2})+$

#matches a word that is reduplicated (catcat, ssss, abab, etc.) 
pats.append('^(\w+)\\1$')

#matches a word with a nasal+stop cluster at the end    
pats.append('^(\w+)[mn][kgtdpb]$') #or: '^(\w+)(m|n)(k|g|t|d|p|b)$'

#matches a monosyllabic word with a stop consonant on each 
#side of the vowel  
#I picked up the monosyllabic words having only a CVC syllable
pats.append('^[kgtdpb][aeoui][kgtdpb]$')   #or: '^(k|g|t|d|p|b)(a|e|o|u|i)(k|g|t|d|p|b)$'

n = int(input('Enter a number (0-4)')) 

i = 0
for word in words:
    if re.search(pats[n],word):
        i += 1
        print(i,word)
    if i > 10:
        break
```

```python
import re
#read the Alice textfile and strip off the header
inFile = open ('alice.txt', 'r')
aliceText = inFile.read()
inFile.close()
aliceText = aliceText[10841:]

#convert everything to lowercase
#I could keep the uppercase letters and use a regex like
#'\s[a-z]*\s([A-Z][a-z]+)\s[a-z]*\s' to find out proper nouns
#but when I did it, I noticed that it captured lots of irrelevant words,
#since I replaced punctuations with spaces,
#so decided not to use this patten 
aliceNewText = aliceText.lower()

#convert all punctuation to spaces
punctuations= ".)-?!`(:,;"  
newAliceText = ""
for c in aliceNewText:
    if c in punctuations:
        newAliceText=newAliceText +''
    else:
        newAliceText= newAliceText+ c
#print (newAliceText)


#split the text by new lines (\n)
aliceLines =newAliceText.split('\n')
#print (aliceLines)

#get rid of empty lines in the list
newAliceLines=[]
for line in aliceLines:
    if line != '':
        newAliceLines.append(line)
#print (newAliceLines)

#define five patterns to be matched 

#pattern 1:, it can overreoprt cases in which
#there is an adjective between the determiner and the noun
pattern1='\s(the|a|an)\s([a-z]+)\s' 

#pattern 2: first \s([a-z]*\s)* was used after possessive pronouns,
#to include cases in which there is an adjective preceding the noun,
#but then it was removed to exclude cases like "your book was ..."
pattern2='\s(my|your|his|her|our|their)\s(own\s)?([a-z]+)\s'

#pattern 3:it can also overreports cases
#in which an adjective precedes the noun 
#like "another long passage, other unpleasant things"
pattern3='\s(other|another)\s([a-z]+)\s'


#pattern 4:removed er/or/ent, 
#since it overreported words such as "for, her, went"
pattern4= '^([a-z]+(tion|ence|ance|sion|ure|ment|age|ness|ity|ism|ant|ship|hood|land)$)'

#pattern 5: noun should have at least 3 letters before the plural 's'
#to exclude cases like "is, as, was, has", 
#but still overreports verbs which have third person singular 's' like 'eats'
pattern5='^([a-z][a-z][a-z][a-z]*(s|es))$'  


#I tried to use 's but it overreported
#the contracted forms of "is" and "has"
#and seemed that the number of these cases were
#far more than the number of nouns with 's 
#pattern6='\s([a-z]+)\'s\s([a-z]+)'

#for each line in the text
for line in newAliceLines[:100]:
    #search for patterns one by one 
    m1=re.search(pattern1,line)
    m2=re.search(pattern2,line)
    m3=re.search(pattern3, line)
    if m1:
        print (m1.group(2))
        
    if m2:
        print(m2.group(3))
    
    if m3:
        print(m3.group(2))
    #for each word in that line    
    for word in line.split():
        if re.search(pattern4, word):
            print (word)
        
        if re.search(pattern5, word):
            print(word)
        
            
            
```
